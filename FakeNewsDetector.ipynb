Import the dependencies

```python
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download('stopwords')
```

```python
print(stopwords.words('english'))
```

Data Pre-Processing

```python
news_dataset = pd.read_csv('/content/drive/MyDrive/Collab data/Fake News Detector/train.csv')
```

```python
news_dataset.shape
```

```python
#print 1st 5 row
news_dataset.head()
```

```python
#missing values
news_dataset.isnull().sum()
```

```python
#replace null value
news_dataset = news_dataset.fillna('')
```

```python
#use only title and author to get result
news_dataset['content'] = news_dataset['author']+' '+news_dataset['title']
```

```python
print(news_dataset['content'])
```

```python
#separate data and label
X = news_dataset.drop(columns='label', axis=1)
Y = news_dataset['label']
```

```python
print(X)
print(Y)
```

```python
#stemming
port_stem = PorterStemmer()
```

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
news_dataset['content'] = news_dataset['content'].apply(stemming)
```

```python
print(news_dataset['content'])
```

```python
#separating the data and label
X = news_dataset['content'].values
Y = news_dataset['label'].values
```

```python
print(X)
print(Y)
```

```python
#convert text in number
vectorizer = TfidfVectorizer()
vectorizer.fit(X)
X = vectorizer.transform(X)
```

```python
print(X)
```

```python
#split x and y for train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
#train the model
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
#Accuracy Score
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy score of the training data is: ", training_data_accuracy)
```

```python
# accuracy score on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy score of the test data is: ", test_data_accuracy)
```

Make a Prediction System

```python
X_new = X_test[2000]

prediction = model.predict(X_new)
print(prediction)

if (prediction[0]==0):
  print('The news is Real.')
else:
  print('The news is Fake.')
```
